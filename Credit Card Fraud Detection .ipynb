```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("creditcard.csv")
```

```python

legit = df[df.Class == 0]
fraud = df[df.Class == 1]
print(legit.shape)
print(fraud.shape)
```

```python
# compare the values for both transactions
df.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)#under_sampling
print(legit_sample)
```

```python
new_df = pd.concat([legit_sample, fraud], axis=0)#Concatenating two DataFrames, axis=0(rows)
new_df.head()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
#Splitting the data into Features & Targets
X = new_df.drop(columns='Class', axis=1)
Y = new_df['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X_train.shape,  X_test.shape, Y_train.shape, Y_test.shape)
```

```python
model = LogisticRegression()
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
y_prediction_train = model.predict(X_train)
training_data_accuracy = accuracy_score(y_prediction_train, Y_train)
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_prediction_train, Y_train))
print(confusion_matrix(y_prediction_train, Y_train))
```

```python
# accuracy on test data
y_prediction_test = model.predict(X_test)
test_data_accuracy = accuracy_score(y_prediction_test, Y_test)
print('Accuracy score on Test Data : ', test_data_accuracy)
      
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_prediction_test, Y_test))
print(confusion_matrix(y_prediction_test, Y_test))
```
