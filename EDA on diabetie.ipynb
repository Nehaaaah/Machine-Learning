```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("diabetes.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data = df, x = "Outcome")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.stripplot(data = df, x = "Outcome" , y = "Age", jitter = True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.swarmplot(data =df, x = "Outcome", y = "Age")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.violinplot(data =df, x = "Outcome", y = "Age")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.violinplot(data =df, x = "Outcome", y = "Age")
sns.swarmplot(data =df, x = "Outcome", y = "Age", color = "black")
plt.show()
```

```python
# point plot , clear comapriosn is possible.
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.pointplot(data =df, x = "Outcome", y = "Age",color="black")
plt.show()
```

```python
# parameter lik aspect , height related to size of subplot. parameter col will give graph in col manner.(side by side)
sns.set(font_scale=1.5)
sns.lmplot(data = df, x = "Age", y = "DiabetesPedigreeFunction", height=8, col="Outcome")
plt.show()
```

```python
df.corr()
```

```python
plt.figure(figsize=(12,8))

sns.set(font_scale=1.4)
sns.heatmap(df.corr(), annot= True, cmap = "Blues")
plt.show()
```

```python
sns.scatterplot(x='Age', y='BMI',
                hue='Outcome', data=df, )

plt.legend()
  
plt.show()
```

```python
#sns.set_style("whitegrid")
sns.pairplot(df, hue="Outcome", size=3)
plt.show()
```

```python
plt.figure(figsize=(10,10))

plt.subplot(221)
plt.title("Glucose")
plt.hist(df['Glucose'])

plt.subplot(222)
plt.title("BloodPressure")
plt.hist(df['BloodPressure'])

plt.subplot(223)
plt.grid()
plt.title("SkinThickness")
plt.hist(df['SkinThickness'])

plt.subplot(224)
plt.grid()
plt.title("Insulin")
plt.hist(df['Insulin'])

plt.show
```

```python
sns.catplot(x="Outcome", y="Glucose", kind="box",fliersize=8.0,data=df)
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

def graph(y):
    sns.boxplot(x="Outcome", y=y, data=df)

plt.figure(figsize=(10,10))

plt.subplot(221)
graph('Age')

plt.subplot(222)
graph("Pregnancies")

plt.subplot(223)
graph("Insulin")

plt.subplot(224)
graph('DiabetesPedigreeFunction')

plt.show()
```
